```python
# Perform setup
import os
import pandas as pd
import matplotlib.pyplot as plt
import re
import json

PROJ_ROOT = os.pardir
```

```python
majorsPath = os.path.join(PROJ_ROOT, "data", "processed", "by_major.csv")
majors = pd.read_csv(majorsPath)
```

```python
majors.head()
```

```python
# Get rid of the first two rows, which contain totals and Undeclared
majors = majors.drop([0,1], axis=0)
```

```python
# Make all names lowercase to get rid of redudancies
majors["Major"] = majors["Major"].map(lambda x: x.lower())
```

```python
majors = majors.groupby("Major").sum().sort_values("Name", ascending=False)
```

```python
# No more indecisive people
majors = majors.drop("undecided", axis=0)
```

```python
majors = majors.reset_index()
```

```python
majors.head()
majors = majors.iloc[:, 0:2]
```

```python
majors.columns = ["Major", "Count"]
majors.head()
govCount = majors.loc[majors["Major"] == "government",]["Count"]
majors.set_value(
    majors["Major"] == "political science",
    "Count",
        majors.loc[majors["Major"] == "political science",]["Count"] + int(govCount)
)
```

```python
majors.head()
```

```python
majors = majors.drop(majors[majors["Major"] == "government"].index, axis=0)
```

```python
majors.head()
```

```python
# Export this table for further visualization
exportPath = os.path.join(PROJ_ROOT, "data", "processed", "majors_processed.csv")
majors.to_csv(exportPath, encoding='utf=8')
```

```python
# Prepare a separate csv containing the top 10 majors
topMajors = majors.iloc[:10,:]
```

```python
topPath = os.path.join(PROJ_ROOT, "data", "processed", "top_majors_processed.csv")
topMajors.to_csv(topPath, encoding='utf=8')
```
